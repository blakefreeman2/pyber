```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_read = pd.read_csv(city_data_to_load)
ride_data_read = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_pyber_df = pd.merge(ride_data_read, city_data_read,how="outer", on='city')

# Display the data table for preview
combined_pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
pyber_group = combined_pyber_df.groupby("city")
pyber_count = pyber_group.type.count()
#urban
urban_city_type = combined_pyber_df.loc[(combined_pyber_df["type"] == 'Urban')]
urban_city_group = urban_city_type.groupby("city")
urban_city_count = urban_city_group.type.count()
urban_city_avg_fare = (urban_city_group['fare'].sum()/urban_city_count)
#suburban                                        
suburban_city_type = combined_pyber_df.loc[(combined_pyber_df["type"] == 'Suburban')]
suburban_city_group = suburban_city_type.groupby("city")
suburban_city_count = suburban_city_group['type'].count()
suburban_city_avg_fare = (suburban_city_group['fare'].sum()/suburban_city_count)
#rural                                
rural_city_type = combined_pyber_df.loc[(combined_pyber_df["type"] == 'Rural')]
rural_city_group = rural_city_type.groupby("city")
rural_city_count = rural_city_group['type'].count()
rural_city_avg_fare = (rural_city_group['fare'].sum()/rural_city_count)
# Build the scatter plots for each city types
urban_city=plt.scatter(urban_city_count, urban_city_avg_fare, marker="o", facecolors="#FF7F50", edgecolors="black",
            s=[10*pyber_count], alpha=0.7, label='urban')

suburban_city=plt.scatter(suburban_city_count, suburban_city_avg_fare, marker="o", facecolors="#ffd700", edgecolors="black",
            s=[10*pyber_count],alpha=0.7,  label='suburban')

rural_city=plt.scatter(rural_city_count, rural_city_avg_fare, marker="o", facecolors="#87CEEB", edgecolors="black",
            s=[10*pyber_count],alpha=0.7, label='rural')
# Incorporate the other graph properties
plt.grid(b=None, which='major', axis='both',linewidth=1)
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
averagefare=plt.ylabel("Average Fare ($)")
# Create a legend
plt.legend(handles=[urban_city, suburban_city,rural_city], title= "City Types", loc="best")
# Incorporate a text label regarding circle size
textstr = 'Note:\nCircle size correlates with driver count per city\n'
plt.text(41.2, 33.5, textstr, fontsize=10)

# Save Figure
plt.savefig("../Images/Ride_Sharing_Data.png")
# Show Figure
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_type = (["Rural", "Suburban","Urban"])
fares_type_group = combined_pyber_df.groupby("type")
fares_type_amount = fares_type_group.fare.sum()
colors = ["#87CEEB","#ffd700","#FF7F50"]
explode = (0,0,0.1)
# Build Pie Chart
plt.pie(fares_type_amount, explode=explode, labels=fares_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140 , pctdistance = .5)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("../Images/Percent_of_Total_Fares_by_City_Type.png")
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_type_count = fares_type_group.fare.count()
colors = ["#87CEEB","#ffd700","#FF7F50"]
explode = (0,0,0.1)
# Build Pie Chart
plt.pie(ride_type_count, explode=explode, labels=fares_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140 , pctdistance = .5)
plt.axis("equal")
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("../Images/Percent_of_Total_Rides_by_City_Type.png")
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_type = city_data_read.groupby("type")
driver_unique = driver_type.driver_count.sum()

colors = ["#87CEEB","#ffd700","#FF7F50"]
explode = (0,0,0.1)
# Build Pie Chart
plt.pie(driver_unique, explode=explode, labels=fares_type, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140 , pctdistance = .5)
plt.axis("equal")
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("../Images/Percent_of_Total_Drivers_by_City_Type.png")
# Show Figure
plt.show()
```
